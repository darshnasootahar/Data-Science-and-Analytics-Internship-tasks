##Exploratory Data Analysis(EDA) with graph 
##Data Visualization
Created the following visualizations:
- Scatter Plot– to analyze relationships between variables.
- Histogram – to examine data distribution.
- Box Plot – to detect outliers and spread of values.

```python
sn.scatterplot(data=fn, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

##histogram

```python
fn['PetalLengthCm'].hist(bins=20)
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

##Box plot 

```python
sn.boxplot(data=fn, x='Species', y='SepalLengthCm')
plt.title('Box Plot of Sepal Length by Species')
plt.show()
```

##Results and Insights
- The Iris dataset contains 150 entries with no missing values.
- Each species shows distinguishable patterns in petal length and width.
- Box plots revealed some mild outliers in sepal width.
- Species can be visually distinguished through scatter plots of petal dimensions.
